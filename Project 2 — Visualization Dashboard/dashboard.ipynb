```python
# monthVScompany=df.groupby(['Month','Company']).agg({
#     'Unnamed': 'count',           
#     'Price ($)': 'sum'                         
# }).round(2)
df['quantity']=1
monthVScompany=df.groupby(['Month','Company']).agg({
    'quantity': 'sum',           
    'Price ($)': 'sum'                         
}).round(2)
```

```python
monthVScompany
```

```python
plt.plot(monthVScompany.Month, monthVScompany.quantity)
```

```python
monthVScompany

```

```python
grouped = df.groupby(['Month', 'Company'])['quantity'].sum().reset_index()
pivot_df = grouped.pivot(index='Month', columns='Company', values='quantity')
grouped
```

```python
pivot_df.plot(kind='line', marker='o', figsize=(12, 6))
plt.title('Monthly Sales Quantity by Car Brand')
plt.xlabel('Month')
plt.ylabel('Quantity')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

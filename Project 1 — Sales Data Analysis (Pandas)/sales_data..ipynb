```python
import pandas as pd
```

```python
import numpy as np
```

```python
df=pd.read_csv('Sales Data.csv')
```

```python
df.head(10)
```

```python
result = df.groupby(['Month', 'City']).agg({
    'Unnamed: 0': 'count',           
    'Price Each': 'sum',             
    'Quantity Ordered': 'sum',       
    'Sales': 'sum'                   
}).round(2)


result.columns = ['total_sales', 'total_price', 'total_quantity', 'total_sales_amount']
result = result.reset_index()
```

```python
result
```

```python
result.to_excel('finalResult.xlsx', sheet_name='Sales Summary', index=False)
```
